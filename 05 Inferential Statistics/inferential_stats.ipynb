### Import the Required Python Packages

```python
import pandas as pd
import numpy as np
import plotly as py

py.offline.init_notebook_mode(connected=True)
```

***
### Import the Dataset(s)

```python
# Import the sugarcane data into a pandas DataFrame
df_sugarcane = pd.read_csv("../03 Data Wrangling/df_sugarcane.csv", header=[0,1], index_col=0)
df_sugarcane.tail()
```

```python
# Import the regional weather data into a pandas DataFrame
df_weather = pd.read_csv("../04 Data Storytelling/df_weather.csv", header=[0,1], index_col=0)
df_weather.tail()
```

***
### Investigate the Distributions

```python
# Inspect the temperature data distributions
print("FL_tmin mean={0:.1f}{4}F and standard deviation={1:.2f}{4}F and FL_tmax mean={2:.1f}{4}F and standard deviation={3:.2f}{4}F" \
      .format(np.mean(df_weather["TMIN"]["FL"]), np.std(df_weather["TMIN"]["FL"]), \
              np.mean(df_weather["TMAX"]["FL"]), np.std(df_weather["TMAX"]["FL"]), u'\xb0'))
print("HI_tmin mean={0:.1f}{4}F and standard deviation={1:.2f}{4}F and HI_tmax mean={2:.1f}{4}F and standard deviation={3:.2f}{4}F" \
      .format(np.mean(df_weather["TMIN"]["HI"]), np.std(df_weather["TMIN"]["HI"]), \
              np.mean(df_weather["TMAX"]["HI"]), np.std(df_weather["TMAX"]["HI"]), u'\xb0'))
print("LA_tmin mean={0:.1f}{4}F and standard deviation={1:.2f}{4}F and LA_tmax mean={2:.1f}{4}F and standard deviation={3:.2f}{4}F" \
      .format(np.mean(df_weather["TMIN"]["LA"]), np.std(df_weather["TMIN"]["LA"]), \
              np.mean(df_weather["TMAX"]["LA"]), np.std(df_weather["TMAX"]["LA"]), u'\xb0'))
print("TX_tmin mean={0:.1f}{4}F and standard deviation={1:.2f}{4}F and TX_tmax mean={2:.1f}{4}F and standard deviation={3:.2f}{4}F" \
      .format(np.mean(df_weather["TMIN"]["TX"]), np.std(df_weather["TMIN"]["TX"]), \
              np.mean(df_weather["TMAX"]["TX"]), np.std(df_weather["TMAX"]["TX"]), u'\xb0'))

trace0 = py.graph_objs.Histogram(name="FL_tmax", x=df_weather["TMAX"]["FL"], histnorm="probability")
trace1 = py.graph_objs.Histogram(name="FL_tmin", x=df_weather["TMIN"]["FL"], histnorm="probability")
trace2 = py.graph_objs.Histogram(name="HI_tmax", x=df_weather["TMAX"]["HI"], histnorm="probability")
trace3 = py.graph_objs.Histogram(name="HI_tmin", x=df_weather["TMIN"]["HI"], histnorm="probability")
trace4 = py.graph_objs.Histogram(name="LA_tmax", x=df_weather["TMAX"]["LA"], histnorm="probability")
trace5 = py.graph_objs.Histogram(name="LA_tmin", x=df_weather["TMIN"]["LA"], histnorm="probability")
trace6 = py.graph_objs.Histogram(name="TX_tmax", x=df_weather["TMAX"]["TX"], histnorm="probability")
trace7 = py.graph_objs.Histogram(name="TX_tmin", x=df_weather["TMIN"]["TX"], histnorm="probability")

print()
fig = py.tools.make_subplots(rows=4, cols=1, shared_xaxes=True, vertical_spacing=0.02)
fig.append_trace(trace0, 1, 1)
fig.append_trace(trace1, 1, 1)
fig.append_trace(trace2, 2, 1)
fig.append_trace(trace3, 2, 1)
fig.append_trace(trace4, 3, 1)
fig.append_trace(trace5, 3, 1)
fig.append_trace(trace6, 4, 1)
fig.append_trace(trace7, 4, 1)

fig["layout"].update(
    plot_bgcolor="rgb(247,247,247)",
    legend=dict(font=dict(family="serif", size=12)),
    height=1000,
    title="<b>Regional Annual Temperature Data Distributions</b>",
    titlefont=dict(family="serif", size=24),
    yaxis1=dict(title="<b>PDF, %</b>", titlefont=dict(family="serif", size=14), tickfont=dict(family="serif", size=14)),
    yaxis2=dict(title="<b>PDF, %</b>", titlefont=dict(family="serif", size=14), tickfont=dict(family="serif", size=14)),
    yaxis3=dict(title="<b>PDF, %</b>", titlefont=dict(family="serif", size=14), tickfont=dict(family="serif", size=14)),
    yaxis4=dict(title="<b>PDF, %</b>", titlefont=dict(family="serif", size=14), tickfont=dict(family="serif", size=14)),
    xaxis1=dict(title="<b>Temperature, {}F</b>".format(u'\xb0'), titlefont=dict(family="serif", size=14),tickfont=dict(family="serif", size=14)))

py.offline.iplot(fig)
```
