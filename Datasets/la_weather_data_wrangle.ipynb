## Combining Weather Station Information for Louisiana
1. Jennings, LA
2. Houma, LA
3. Baton Rouge LSU, LA

```python
# Reshape the historical weather data for Jennings, LA to extract only the relavent data
# for future analysis
df_la_j = df_la_jn[["DATE", "PRCP", "SNOW", "TMAX", "TMIN", "WDMV"]].sort_values("DATE")
df_la_j.DATE = pd.to_datetime(df_la_j.DATE)
df_la_j = df_la_j.set_index("DATE")
df_la_j["Year"] = df_la_j.index.year
df_la_j["Month"] = df_la_j.index.month
df_la_j["YYYY-MM"] = df_la_j.index.strftime("%Y-%m")

# Reshape the historical weather data for Houma, LA to extract only the relavent data
# for future analysis
df_la_h = df_la_hm[["DATE", "PRCP", "SNOW", "TMAX", "TMIN", "WDMV"]].sort_values("DATE")
df_la_h.DATE = pd.to_datetime(df_la_h.DATE)
df_la_h = df_la_h.set_index("DATE")
df_la_h["Year"] = df_la_h.index.year
df_la_h["Month"] = df_la_h.index.month
df_la_h["YYYY-MM"] = df_la_h.index.strftime("%Y-%m")

# Reshape the historical weather data for Baton Rouge LSU, LA to extract only the relavent data
# for future analysis
df_la_b = df_la_brlsu[["DATE", "PRCP", "SNOW", "TMAX", "TMIN", "WDMV"]].sort_values("DATE")
df_la_b.DATE = pd.to_datetime(df_la_b.DATE)
df_la_b = df_la_b.set_index("DATE")
df_la_b["Year"] = df_la_b.index.year
df_la_b["Month"] = df_la_b.index.month
df_la_b["YYYY-MM"] = df_la_b.index.strftime("%Y-%m")

# Data reshaped/resampled to be indexed by YYYY-MM with the appropriate aggregation method
df_la_j_lines_wind = df_la_j.pivot_table(index="YYYY-MM", values="WDMV", aggfunc="mean").dropna()
df_la_h_lines_wind = df_la_h.pivot_table(index="YYYY-MM", values="WDMV", aggfunc="mean").dropna()
df_la_b_lines_wind = df_la_b.pivot_table(index="YYYY-MM", values="WDMV", aggfunc="mean").dropna()
```

```python
# Plot the historical weather combination Dataframe candidates for some quick Exploratory
# Data Analysis (EDA)
# Generate an interactive plot using the plotly package

trace1 = go.Scatter(
    x=df_la_j_lines_wind.index,
    y=df_la_j_lines_wind.WDMV,
    mode="lines",
    line=dict(color="rgb(35,122,181)"),
    name="Jennings, LA")

trace2 = go.Scatter(
    x=df_la_h_lines_wind.index,
    y=df_la_h_lines_wind.WDMV,
    mode="lines",
    line=dict(color="rgb(255,127,14)"),
    name="Houma, LA")

trace3 = go.Scatter(
    x=df_la_b_lines_wind.index,
    y=df_la_b_lines_wind.WDMV,
    mode="lines",
    line=dict(color="rgb(44,160,44)"),
    name="LSU, LA")

trace4 = go.Box(
    y=df_la_j_lines_wind.WDMV,
    marker=dict(color="rgb(35,122,181)"),
    boxmean="sd",
    name="Jennings, LA"
)

trace5 = go.Box(
    y=df_la_h_lines_wind.WDMV,
    marker=dict(color="rgb(255,127,14)"),
    boxmean="sd",
    name="Houma, LA"
)

trace6 = go.Box(
    y=df_la_b_lines_wind.WDMV,
    marker=dict(color="rgb(44,160,44)"),
    boxmean="sd",
    name="LSU, LA"
)

fig = tools.make_subplots(rows=3, cols=2, shared_xaxes=True, shared_yaxes=True,
                          vertical_spacing=0.01)
fig.append_trace(trace1, 1, 1)
fig.append_trace(trace2, 2, 1)
fig.append_trace(trace3, 3, 1)
fig.append_trace(trace4, 1, 2)
fig.append_trace(trace5, 2, 2)
fig.append_trace(trace6, 3, 2)

fig["layout"].update(
    plot_bgcolor="rgb(247,247,247)",
    legend=dict(font=dict(family="serif", size=12)),
    height=1000,
    title="<b>Historical Wind Data for Louisiana Station Combination Candidates</b>",
    titlefont=dict(family="serif", size=24),
    yaxis1=dict(titlefont=dict(family="serif", size=14),
                tickfont=dict(family="serif", size=14)),
    yaxis2=dict(title="<b>Mean Wind Movement, miles</b>",
                titlefont=dict(family="serif", size=14),
                tickfont=dict(family="serif", size=14),
                anchor="x"),
    yaxis3=dict(titlefont=dict(family="serif", size=14),
                tickfont=dict(family="serif", size=14)),
    yaxis4=dict(titlefont=dict(family="serif", size=14),
                tickfont=dict(family="serif", size=14),
                anchor="x2"),
    yaxis5=dict(titlefont=dict(family="serif", size=14),
                tickfont=dict(family="serif", size=14)),
    yaxis6=dict(titlefont=dict(family="serif", size=14),
                tickfont=dict(family="serif", size=14)),
    xaxis1=dict(title="<b>Year</b>", titlefont=dict(family="serif", size=14), 
                tickfont=dict(family="serif", size=14),
                domain=[0, 0.85]),
    xaxis2=dict(tickfont=dict(family="serif", size=14),
                domain=[0.9, 1]),
    xaxis3=dict(title="<b>Year</b>", titlefont=dict(family="serif", size=14),
                tickfont=dict(family="serif", size=14),
                domain=[0, 0.85]),
    xaxis4=dict(tickfont=dict(family="serif", size=14),
                domain=[0.9, 1]),
    xaxis5=dict(title="<b>Year</b>", titlefont=dict(family="serif", size=14), 
                tickfont=dict(family="serif", size=14),
                domain=[0, 0.85]),
    xaxis6=dict(tickfont=dict(family="serif", size=14),
                domain=[0.9, 1]))
py.offline.iplot(fig)
```
